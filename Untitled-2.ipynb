## FOR LOOP

```python
foods = ["cake", "ice cream", "cookies", "carbonara"]

for food in foods:
    print(f"I like eating {food}")
```

```python
for i in range(8):
    print(i)
```

```python
user = {"name": "Glenn", "age": 21, "city": "Bacolod"}

for key in user:
    print(key)

for val in user.values():
    print(val)
```

```python
for key, value in user.items():
    print(key, ":", value)
```
