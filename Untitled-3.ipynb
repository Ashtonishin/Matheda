## NUMPY FOR MATHEDA

```python
import numpy as np
from scipy import stats

Jaguar = [10.6, 9.1, 9.5, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 9.5, 11, 10.4, 3, 9.8]

Panther = [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1]
def calculate_statistics(name, data):
    print(f"Statistics for {name}:")
    print(f"Mean: {np.mean(data)}")
    print(f"Median: {np.median(data)}")

    mode_result = stats.mode(data, keepdims=True)
    print(f"Mode: {mode_result.mode[0]} (Count: {mode_result.count[0]})")
    print()
calculate_statistics( "Jaguar", Jaguar)
calculate_statistics( "Panther", Panther)
```
